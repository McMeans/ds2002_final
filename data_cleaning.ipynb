# Notebook to clean imported data

```python
import pandas as pd

# load data
df = pd.read_csv('data/13k-recipes.csv')
df
```

```python
# keep only the recipe name, ingredients (cleaned), and instructions
df = df[['Title', 'Cleaned_Ingredients', 'Instructions']]

# rename ingredients column for simplicity
df = df.rename(columns={"Cleaned_Ingredients": "Ingredients"})

# make column names lowercase for simplicity
df.columns = df.columns.str.lower()
df
```

```python
# save as CSV and JSON
df.to_csv('./data/cleaned_recipes.csv', index=False)
df.to_json('./data/cleaned_recipes.json', orient='records')
```

```python
# save as SQL
import sqlite3
conn = sqlite3.connect('./data/cleaned_recipes.db')
df.to_sql('cleaned_recipes', conn, if_exists='replace', index=False)
conn.close()
```
